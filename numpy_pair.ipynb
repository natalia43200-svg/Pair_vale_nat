1. Crea un array 1D con 50 valores aleatorio del 10 al 99. Accede a los elementos en las posiciones múltiplos de 5.

```python
a1d = np.random.randint(10,99,50)

print(a1d)
```

```python
len(a1d)
```

```python
a1d[:5]
```

```python
lista_multiplos_5 = []


for n in range(len(a1d)): 
    if n%5 == 0:
        lista_multiplos_5.append(n)

print(lista_multiplos_5)
```

```python
lista_multiplos_5 = []


for n in a1d: 
    if n%5 == 0:
        lista_multiplos_5.append(n)

print(lista_multiplos_5)
```

- 2. Crea un array 1D con 30 valores aleatorios del 1 al 20 (ambos incluidos). Utiliza indexación para obtener los elementos en las posiciones impares.

```python
a2_1d = np.random.randint(1,22,30)

print(a2_1d)
```

```python
lista_impares = []


for n in a2_1d:
    if n%2 != 0:
        lista_impares.append(n)

print(lista_impares)
```
