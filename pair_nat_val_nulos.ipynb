## Informe de Limpieza de Datos - Tratamiento de Valores Nulos

### Columnas con Valores Nulos Identificadas
El análisis inicial identificó **27 columnas numéricas** con valores faltantes en el dataset `world_data_full_apply.csv`. Las principales fueron: densidad, tierra agrícola, emisiones de CO2, PIB, esperanza de vida, tasa de mortalidad infantil, entre otras.

### Método de Imputación Seleccionado
Se utilizó el método **KNN Imputer (K-Nearest Neighbors)** con 5 vecinos más cercanos para imputar los valores nulos. Esta técnica fue elegida por:
- **Ventaja**: Captura relaciones complejas entre variables, utilizando similitud entre países basada en características existentes.
- **Contexto**: Apropiado para datos geográficos/socioeconómicos donde países similares tienden a valores similares.
- **Robustez**: Mejor que media/mediana para datasets heterogéneos con múltiples dimensiones.

### Observaciones Importantes del Proceso

1. **Preservación de Estructura**: La imputación mantiene la distribución y correlaciones entre variables mejor que métodos simples.

2. **Validación**: Después de la imputación, `df.isnull().sum()` confirma 0 valores faltantes en todas las columnas tratadas.

3. **Limitaciones**: El KNN asume que valores similares se encuentran cercanos en el espacio de características. En casos de valores extremos aislados, podría suavizar demasiado las imputaciones.

4. **Siguientes Pasos**: Se recomienda explorar visualizaciones (boxplots, histogramas) para validar que las distribuciones imputadas son coherentes con los datos originales y realizar análisis de sensibilidad si el modelo es crítico.

 # METODOS:
 - Numérico (ej. edad, ingresos): Media (si distribución normal) o mediana (si sesgada).
- Categórico (ej. género, país): Moda (valor más frecuente).
- Avanzado: KNNImputer para patrones complejos o IterativeImputer para relaciones entre variables.
