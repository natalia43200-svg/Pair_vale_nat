# UNION:

    - concact -> lo usamos para poner un df debajop de otro (axis=0)
    - join -> une los DFs uno al lado de otro -> une por los índices
    - merge -> usamos una columna de union, similar a lo que conocemos en SQL -->

```python
df_unido_concact = pd.concat([df_part1, df_part2], axis =0, ignore_index= True)
df_unido_concact.head()
```

```python
df_unido_join = df_part1.join(df_part2)
df_unido_join.head()
```

```python
df_unido_merge = df_part1.merge(df_part2, left_on="Country", right_on="country")
df_unido_merge.head()
```

- Nos quedamos con el metodo Merge porque tenemos las columnas de country que nos coinciden

```python
df_unido_merge.columns 
```

```python

# Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

# No tengan espacios.

# Estén en minúscula.

# No tengan paréntesis, es decir, quitar "(%)", "(Km2)".

# Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

# Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.
```

```python
df_unido_merge.columns = df_unido_merge.columns.str.strip().str.lower().str.replace(r'\n', '').str.replace(r'\([^)]*\)', '').str.replace(':', '').str.replace(' ', '')
```

```python
df_unido_merge.columns = (df_unido_merge.columns).str.strip() #le hemos quitado los espacios INVISIBLES
```

```python
df_unido_merge.columns = (df_unido_merge.columns).str.replace(' ', '') #le hemos quitado los espacios VISIBLES
```

```python
df_unido_merge.columns = (df_unido_merge.columns).str.lower() #le hemos quitado las minusculas
```

```python
df_unido_merge.columns = (df_unido_merge.columns).str.replace("(%)", "") 
```

```python
df_unido_merge.columns = (df_unido_merge.columns).str.replace( "(Km2)", "")
```

```python
df_unido_merge.columns = (df_unido_merge.columns).str.replace( "\n", "")
```

```python
df_unido_merge.columns
```

```python
df_unido_merge["coordinates"]
```

```python
cols = df_unido_merge['coordinates'].astype(str).str.strip('() ').str.split(',', expand=True)
df_unido_merge['longitud'] = cols[0]
df_unido_merge['latitud'] = cols[1]
```

```python
df_unido_merge
```

```python
df_unido_merge.drop(columns=["coordinates"], inplace=True) 
```

```python
df_unido_merge
```

```python
#borrar los % de los valores dentro de las columnas
# unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land
```

```python
# 1. Identificamos las columnas que son de tipo texto (object)
# Esto evita que intentemos limpiar columnas que ya son números
columnas_texto = df_unido_merge.select_dtypes(include=['object']).columns

# 2. Aplicamos la limpieza solo a esas columnas
for col in columnas_texto:
    # Verificamos si el símbolo existe en la columna para no dar errores
    if df_unido_merge[col].astype(str).str.contains('%').any():
        df_unido_merge[col] = df_unido_merge[col].str.replace('%', '', regex=False).astype(float)

print("¡Columnas limpias y convertidas a números!")
```

Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".



```python
# 1. Identificamos las columnas que son de tipo texto (object)
# Esto evita que intentemos limpiar columnas que ya son números
columnas_texto = df_unido_merge.select_dtypes(include=['int']).columns

# 2. Aplicamos la limpieza solo a esas columnas
for col in columnas_texto:
    # Verificamos si el símbolo existe en la columna para no dar errores
    if df_unido_merge[col].astype(str).str.contains('$').any():
        df_unido_merge[col] = df_unido_merge[col].str.replace('$', '', regex=False)
```
