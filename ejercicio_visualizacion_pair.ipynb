```python
import seaborn as sns
import matplotlib.pyplot as plt

europa = df[df["continent"] == "Europe"]

sns.boxplot(y=europa["life_expectancy"])
plt.title("Esperanza de vida en Europa")
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 6))

sns.scatterplot(data=df, x="birth_rate", y="infant_mortality", ax=ax[0,0])
sns.scatterplot(data=df, x="birth_rate", y="life_expectancy", ax=ax[0,1])
sns.scatterplot(data=df, x="birth_rate", y="density", ax=ax[1,0])
sns.scatterplot(data=df, x="birth_rate", y="urban_population", ax=ax[1,1])

plt.tight_layout()
plt.show()
```

Relacion entre  infant moratality y birth rath es una correlacion positiva , life expectancy y birth rate correlacion negativa , density no tiene relacion aparente ni urban population con birth rate
